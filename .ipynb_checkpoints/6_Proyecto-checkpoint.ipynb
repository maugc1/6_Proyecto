### Test

```python
test.salary_currency.unique()
```

```python
test.head()
```

```python
test['salary_usd']=0
```

```python
for i in range(len(test.salary)):
    if test.salary_currency[i]=='USD':
        test['salary_usd'][i]= test.salary[i]*1
    elif test.salary_currency[i]=='CNY':
        test['salary_usd'][i]= test.salary[i]*0.15   
    elif test.salary_currency[i]=='INR':
        test['salary_usd'][i]= test.salary[i]*0.013   
    elif test.salary_currency[i]=='MXN':
        test['salary_usd'][i]= test.salary[i]*0.049
    elif test.salary_currency[i]=='EUR':
        test['salary_usd'][i]= test.salary[i]*0.99
    elif test.salary_currency[i]=='DKK':
        test['salary_usd'][i]= test.salary[i]*0.14
    elif test.salary_currency[i]=='HUF':
        test['salary_usd'][i]= test.salary[i]*0.0025   
    elif test.salary_currency[i]=='GBP':
        test['salary_usd'][i]= test.salary[i]*1.20
    elif test.salary_currency[i]=='CAD':
        test['salary_usd'][i]= test.salary[i]*0.77
    elif test.salary_currency[i]=='SGD':
        test['salary_usd'][i]= test.salary[i]*0.72
    elif test.salary_currency[i]=='BRL':
        test['salary_usd'][i]= test.salary[i]*0.19
    elif test.salary_currency[i]=='PLN':
        test['salary_usd'][i]= test.salary[i]*0.21   
```

```python
test.head()
```

```python
plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(test.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(test.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```

```python
test.drop('work_year',axis=1,inplace=True)
test.drop('remote_ratio',axis=1,inplace=True)
test.drop('salary_currency',axis=1,inplace=True)
test.drop('salary',axis=1,inplace=True)
```

```python
test.head()
```

```python
X_test=test.drop('salary_usd', axis=1)

y_test=test.salary_usd

X_test.head()
```

```python
from sklearn.preprocessing import LabelEncoder

X_test.experience_level=LabelEncoder().fit_transform(X_test.experience_level)
X_test.employment_type=LabelEncoder().fit_transform(X_test.employment_type)
X_test.job_title=LabelEncoder().fit_transform(X_test.job_title)
X_test.employee_residence=LabelEncoder().fit_transform(X_test.employee_residence)
X_test.company_location=LabelEncoder().fit_transform(X_test.company_location)
X_test.company_size=LabelEncoder().fit_transform(X_test.company_size)
X_test.head()
```
